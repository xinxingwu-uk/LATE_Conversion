```python
import pandas as pd
import numpy as np
import gc
```

```python
def csv_read(p_path):
    pd_data=pd.read_csv(p_path, sep=',')
    column_names=pd_data.columns.values.tolist()
    print(column_names)
    return pd_data,np.array(pd_data),column_names

def write_to_csv(p_data,p_path):
    dataframe = pd.DataFrame(p_data)
    dataframe.to_csv(p_path, mode='a',header=False,index=False,sep=',')
    del dataframe
    gc.collect()
```

```python
data,data_arr,column_names=csv_read('./data/survival_data_filtered_2.csv')
```

```python
column_names_del=[]
for col in np.arange(data_arr.shape[1]):
    if np.sum(data_arr[:,col]<0)>0:        
        column_names_del.append(column_names[col])
```

```python
column_names_del
```

```python
data_preprocess=data.drop(columns=column_names_del)
```

```python
data_preprocess.shape
```

```python
data_preprocess
```

```python
data_preprocess_arr=np.array(data_preprocess)
data_store_title=data_preprocess.columns.values.tolist()
```

```python
len(data_store_title)
```

```python
CDRGLOB_index=np.where(np.array(data_store_title)=='CDRGLOB')[0][0]
CDRGLOB_index
```

```python
Event_index=np.where(np.array(data_store_title)=='Event')[0][0]
Event_index
```

```python
del_non_change_status=np.where((data_preprocess_arr[:,CDRGLOB_index]!=0.5) | (data_preprocess_arr[:,Event_index]!=False))[0]
```

```python
data_preprocess
```

```python
data_preprocess_drop=data_preprocess.drop('CDRGLOB', axis=1)
data_store_title_drop=data_preprocess_drop.columns.values.tolist()

data_preprocess_drop
```

```python
len(data_store_title_drop)
```

```python
import matplotlib.pyplot as plt
import numpy as np

title=np.array(data_store_title_drop)
data_preprocess_drop_arr=np.array(data_preprocess_drop)[del_non_change_status]

TrueSamples=np.sum(data_preprocess_drop_arr[:,-1]==True)
print("True Samples:", TrueSamples)

FalseSamples=np.sum(data_preprocess_drop_arr[:,-1]==False)
print("False Samples:", FalseSamples)

MaxTF_Samples=max(TrueSamples,FalseSamples)

fz=20
params = {
    'figure.figsize': '8, 6'
}
plt.rcParams.update(params)

labels=[r'{0}$\rightarrow${0.5}',r'{0}$\nrightarrow${0.5}']
data=[TrueSamples,FalseSamples]
width=0.3
rects_1=plt.bar(2,data[0],width,label=labels[0],color='#d5695d')
height = rects_1[0].get_height()
plt.text(rects_1[0].get_x() + rects_1[0].get_width() / 2, height, str(height), size=fz, ha='center', va='bottom')

rects_2=plt.bar(3,data[1],width,label=labels[1],color='#5d8ca8')

height = rects_2[0].get_height()
plt.text(rects_2[0].get_x() + rects_2[0].get_width() / 2, height, str(height), size=fz, ha='center', va='bottom')

plt.title('')
plt.xlim(1.2,3.8)
plt.ylim(0,MaxTF_Samples)
plt.yticks(np.arange(0,MaxTF_Samples+40,50),fontsize=fz)
plt.xticks([1.5,2,2.5,3,3.5],['',labels[0],'',labels[1],''],fontsize=fz)
plt.xlabel('')
plt.ylabel('Number of samples',fontsize=fz)
plt.show()
```

```python
print("%.2f%% samples are right censored in whole data." % (np.sum(data_preprocess_drop_arr[:,-1]==False) * 100. / len(data_preprocess_drop_arr[:,-1])))
```
