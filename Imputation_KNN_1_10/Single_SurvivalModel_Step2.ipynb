```python
import pandas as pd
import numpy as np
import gc
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import KNNImputer
```

```python
def csv_read(p_path):
    pd_data=pd.read_csv(p_path, sep=',')
    column_names=pd_data.columns.values.tolist()
    print(column_names)
    return pd_data,np.array(pd_data),column_names

def write_to_csv(p_data,p_path):
    dataframe = pd.DataFrame(p_data)
    dataframe.to_csv(p_path, mode='a',header=False,index=False,sep=',')
    del dataframe
    gc.collect()
```

```python
data,data_arr,column_names=csv_read('./data/survival_data_filtered_2.csv')
```

```python
column_names_del=[]
for col in np.arange(data_arr.shape[1]):
    if np.sum(data_arr[:,col]<0)>10:        
        column_names_del.append(column_names[col])
```

```python
data_preprocess_=data.drop(columns=column_names_del)
```

```python
data_preprocess_[data_preprocess_ < 0] = np.nan
```

```python
data_preprocess_doing = data_preprocess_.apply(lambda series: pd.Series(
        LabelEncoder().fit_transform(series[series.notnull()]),
        index=series[series.notnull()].index
    ))
    
imputer = KNNImputer()
data_preprocess_imputer=imputer.fit_transform(data_preprocess_doing).astype(int).astype(object)
```
